```python
from matplotlib.artist import Artist 
import matplotlib  
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import pandas as pd
import numpy as np
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(x[i],y[i]+1.5,y[i], **csfont)
```

```python
df = pd.read_excel('Homeowner_race.xlsx').rename(columns={"Unnamed: 0": "year"})
df2 = pd.read_excel('Homeowner_race_1.xlsm').rename(columns={"Unnamed: 0": "year"})
df2
```

```python
fig, ax = plt.subplots(figsize=(8,6),dpi=120)  
# font = FontProperties()
# font.set_family('cursive')

csfont = {'fontname':'serif'}
ticks_font = FontProperties(family='serif', style='normal',
    weight='normal', stretch='normal')
plt.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# plt.text(2010, 30, 30, size=12, math_fontfamily='cm')

ax.plot(df.year, df.White, color='blue')
ax.scatter(df.year, df.White, marker='o', facecolors='none', edgecolor='blue')
ax.plot(df.year, df.Asian, color='lightgreen')
ax.scatter(df.year, df.Asian, marker='o', facecolors='none', edgecolor='lightgreen')
ax.set_title('United States owner-occupied housing percentage by Race', loc='left', y=0.92, x=0.02, fontsize='medium', **csfont)
plt.rcParams['lines.linewidth'] = 2

Artist.set(ax, xlim =(2009, 2023), ylim =(0.0, 100.0))

for label in ax.get_xticklabels():
    label.set_fontproperties(ticks_font)

addlabels(np.array(df.year), np.array(df.White))
addlabels(np.array(df.year), np.array(df.Asian))
plt.ylabel('Percentage', fontsize=14, **csfont)
ax.grid(alpha=0.3)
ax.legend(['White', 'Asian'], loc='center right', prop={'family': 'serif'})
plt.figtext(0.12, 0.05, "Data Sources:", weight='bold', **csfont)
plt.figtext(0.27, 0.05, "U.S. Census Bureau, 2010-2022 American Community Survey PUMS;")
plt.figtext(0.27, 0.02, "Federal Financial Institutions Examination Council")
# plt.show()
plt.savefig('US_housing.png')
```
